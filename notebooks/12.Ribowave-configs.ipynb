```python
sheet_name = 'hg38'
excel_df = pd.read_excel('../data/datasets_info.xlsx', sheet_name=sheet_name)
config_out_dir = '/auto/cmb-panasas2/skchoudh/github_projects/ribocop-results/tools/Ribowave/configs/'
input_dir_root = '/home/cmb-panasas2/wenzhenl/benchmark/gene_level/hg38/'
output_dir_root = '/staging/as/skchoudh/benchmark/gene_level/hg38/'

genome_fasta = '/home/cmb-panasas2/skchoudh/genomes/hg38/fasta/hg38.fa'
gtf = '/home/cmb-panasas2/skchoudh/genomes/hg38/annotation/gencode.v25.annotation.gtf'
ribowave_annotation_dir = '/home/cmb-panasas2/skchoudh/genomes/hg38/ribowave_annotation'
chrom_sizes = '/home/cmb-panasas2/skchoudh/genomes/hg38/fasta/hg38.chrom.sizes'

for index, row in excel_df.iterrows():
    lengths = list(row['read lengths'].strip().split(','))
    offsets = row['offsets'].strip().split(',')
    srp = row['SRA']
    ribo_sample = row['RIBO sample']
    rna_sample = row['RIBO sample']
    input_dir = os.path.join(input_dir_root, srp+'_human')
    output_dir = os.path.join(output_dir_root, srp+'_human')
    snakemake_config = '''
    GTF = '{}'
    CHROM_SIZES = '{}'
    GENOME_FASTA = '{}'
    RIBOWAVE_ANNOTATION_DIR = '{}'
    INPUT_DIR = '{}'
    OUT_DIR = '{}'
    LENGTHS = {}
    OFFSETS = {}
    SAMPLES = {}
    '''.format(gtf, chrom_sizes, genome_fasta, ribowave_annotation_dir,
           input_dir, output_dir, lengths, offsets, 
           ['RNA_CCDS_lengths', 'RIBO_CCDS_lengths'])
    #snakemake_config = snakemake_config.replace("'", '')
    snakemake_config_file = os.path.join(config_out_dir, '{}_human.py'.format(srp))
    with open(snakemake_config_file, 'w') as fh:
        fh.write(dedent(snakemake_config))
```
