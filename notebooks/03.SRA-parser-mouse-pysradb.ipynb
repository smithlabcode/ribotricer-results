```python
%pylab inline
from collections import OrderedDict
from pysradb import SRAdb, expand_sample_attribute_columns
import pandas as pd
```

```python
db = SRAdb('/staging/as/skchoudh/SRAmetadb.sqlite')
```

```python
assembly_common_name_map = {'mm10': 'mouse', 'hg38': 'human'}

sheet_name = 'mm10'
excel_df = pd.read_excel('../data/datasets_info.xlsx', sheet_name=sheet_name)
```

```python
excel_df
```

```python
all_dfs =  OrderedDict()
```

```python
srp = 'SRP003554'
df = db.sra_metadata(srp)
df['assay'] = 'rna'
df = expand_sample_attribute_columns(df)
df.loc[df.experiment_title.str.contains('Footprint'), 'assay'] = 'ribo'
df.to_csv('./results/{}_runtable.tsv'.format(srp), header=True, index=False, sep='\t')
all_dfs[srp] = df
df
```

```python
writer = pd.ExcelWriter('./results/{}_datasets_runtable.xlsx'.format(assembly_common_name_map[sheet_name]), engine='xlsxwriter')
for sheet_name, df in all_dfs.items():
    df.to_excel(writer, sheet_name=sheet_name, index=False)
writer.save()
```
