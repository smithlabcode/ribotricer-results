```python
%pylab inline
import os
import pandas as pd
from textwrap import dedent
```

# Human

```python
assembly_common_name_map = {'mm10': 'mouse', 'hg38': 'human'}

sheet_name = 'hg38'
excel_df = pd.read_excel('../data/datasets_info.xlsx', sheet_name=sheet_name)
excel_df
```

```python
config_out_dir = '/auto/cmb-panasas2/skchoudh/github_projects/ribocop-results/tools/riboHMM/configs'
synfoot_prefix = '/staging/as/skchoudh/riboHMM_synthetic_footprints/hg38/hg38_synfootprints'
input_dir_root = '/home/cmb-panasas2/skchoudh/benchmark/gene_level/hg38/'
intron_bed = '/home/cmb-panasas2/skchoudh/genomes/hg38/annotation/gencode.v25.gffutils.intron.bed'
chrom_sizes = '/home/cmb-panasas2/skchoudh/genomes/hg38/fasta/hg38.chrom.sizes'
genome_fasta = '/home/cmb-panasas2/skchoudh/genomes/hg38/fasta/hg38.fa'
gtf = '/home/cmb-panasas2/skchoudh/genomes/hg38/annotation/gencode.v25.annotation.gtf'
star_index = '/home/cmb-panasas2/skchoudh/genomes/hg38/star_annotated/'
```

```python
offset_file_map = {}
for index, row in excel_df.iterrows():
    read_lengths = list(row['read lengths'].strip().split(','))
    offsets = row['offsets'].strip().split(',')
    srp = row['SRA']
    ribo_sample = row['RIBO sample']
    rna_sample = row['RIBO sample']
    input_dir = os.path.join(input_dir_root, srp+'_human')
    config = dedent('''
    INPUT_DIR = '{}'
    SYNFOOTPRINTS_PREFIX = '{}'
    READ_LENGTHS = {}
    INTRON_BED = '{}'
    CHROM_SIZES = '{}'
    GTF = '{}'
    GENOME_FASTA = '{}'
    STAR_INDEX = '{}'
    ''').format(input_dir, synfoot_prefix, read_lengths, intron_bed, chrom_sizes, gtf, genome_fasta, star_index)
    with open(os.path.join(config_out_dir, '{}_human.py'.format(srp)), 'w') as fh:
        fh.write(config)
    
```
